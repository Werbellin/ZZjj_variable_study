```python
def get_efficieny(file_name, plot_name) :
    teff = get_copy(prefix + file_name, 'raw/' + plot_name)
    total = teff.GetTotalHistogram().Integral()
    passed  = teff.GetPassedHistogram().Integral()

    #print 'passed ', passed
    #print 'total ', total
    return passed / total 
```
