```python
import sys

sys.path.append('/Users/pigard/CMS/ZZ_analysis')

from plotefficiencies import *
import ROOT
```

```python
prefix = '/Users/pigard/CMS/ZZ_analysis/'


def plot_efficiencies(file_name, plot_name, y_title) :
    plot = save_graphs(simple_plots =  [
          (get_graph(prefix + file_name, 'raw/' + plot_name + '__all'), 'pp > zz > 4l', {'legend_title' : 'sum', 'color' : ROOT.kBlack, 'MarkerStyle': 27}),
          (get_graph(prefix + file_name, 'raw/' + plot_name + '__4m'), 'pp > zz > 4l', {'legend_title' : '4#mu', 'color' : ROOT.kRed, 'MarkerStyle': 27}),
          (get_graph(prefix + file_name, 'raw/' + plot_name + '__2e2m'), 'pp > zz > 4l', {'legend_title' : '2e2#mu', 'color' : ROOT.kBlue, 'MarkerStyle': 27}),
          (get_graph(prefix + file_name, 'raw/' + plot_name + '__4e'), 'pp > zz > 4l', {'legend_title' : '4e', 'color' : ROOT.kGreen, 'MarkerStyle': 27})
                            ],
                           save_name = 'eff/' + plot_name, y_title = y_title, draw_options = {})
```
